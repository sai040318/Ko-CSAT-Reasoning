```python
import pandas as pd, ast

df = pd.read_csv("/data/ephemeral/home/sehun/pro-nlp-generationfornlp-nlp-09/output/history_with_context_eval.csv")
# problems 파싱
problems = df["problems"].apply(ast.literal_eval)
df["question"] = problems.apply(lambda x: x["question"])
df["choices"] = problems.apply(lambda x: x["choices"])
df["answer"] = problems.apply(lambda x: int(x["answer"]))

# 오답만 필터

wrong = df[df["answer"] != df["predict"]].copy()  # prediction 컬럼 이름은 파일 확인 후 맞춰주세요
print(len(wrong), "wrong samples")
display(wrong[["id", "question", "choices", "answer", "predict", "paragraph"]].head())

# 필요하면 context만 따로 보기
wrong_ctx = wrong[["id", "paragraph"]]  # paragraph 맨 끝에 [CONTEXT]가 붙어 있음
```

```python
# 텍스트 잘림 방지 (전문 출력)
pd.set_option('display.max_colwidth', None)
# 열 개수 제한 해제
pd.set_option('display.max_columns', None)
# 행 개수 100개까지 출력
pd.set_option('display.max_rows', 100)
```

```python
# 정답 인덱스가 1부터 시작한다고 가정 (데이터셋 포맷에 따라 0이면 -1 제거 필요)
def get_choice_text(choices, idx):
    try:
        return choices[int(idx) - 1] # 1-based index인 경우
    except:
        return "Unknown"

print(f"총 오답 개수: {len(wrong)}개\n" + "="*60)

for i, row in wrong.iterrows():
    print(f"🆔 ID: {row['id']}")
    print(f"❓ 질문: {row['question']}")
    print("-" * 30)
    
    # 선택지 출력
    for idx, choice in enumerate(row['choices']):
        prefix = "✅" if (idx + 1) == row['answer'] else "❌" if (idx + 1) == row['predict'] else "  "
        print(f"{prefix} {idx+1}. {choice}")
    
    print("-" * 30)
    print(f"🎯 정답: {row['answer']}번 ({get_choice_text(row['choices'], row['answer'])})")
    print(f"🤖 예측: {row['predict']}번 ({get_choice_text(row['choices'], row['predict'])})")
    print("-" * 30)
    print(f"📄 [참고 문서(Context)]:\n{row['paragraph']}")
    print("=" * 60 + "\n")
```
